```python
import pandas as pd
import numpy as np

def padding(x):
    x = str(x)
    return x.rjust(6, "0")
```

```python
df_전처리끝 = pd.read_csv("./datasets/df_부실징후_전처리끝.csv", index_col=0)
df_전처리끝["거래소코드"] = df_전처리끝["거래소코드"].map(padding)
df_전처리끝
```

```python
df_실제회생기업 = df_전처리끝[df_전처리끝["회생여부"]==1][["회사명","거래소코드","회계년도","기업생명주기"]].sort_values(by="회계년도", ascending=True)
df_실제회생기업
```

```python
df_실제회생기업.to_csv("./datasets/실제회생기업.csv")
```
