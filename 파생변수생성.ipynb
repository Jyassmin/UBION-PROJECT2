```python
import pandas as pd
import numpy as np
from numpy import log as ln
```

```python
# func 모음

# 1. 종목코드를 6자리로 통일. ex) 990 -> 000990
def padding(x):
    x = str(x)
    return x.rjust(6, "0")

# 2. 회계년도컬럼, 년도만 뽑아서 int화
def change_year(x):
    return int(x[:4])
```

```python
df = pd.read_csv("./datasets/raw/사업보고서/사업보고서_통합.csv", index_col=0)
df.reset_index(inplace=True, drop=True)
df["거래소코드"] = df["거래소코드"].map(padding)
df
```

```python
# pd.set_option('display.max_rows', None)
# df.head()
```

```python
df['배당금합산'] = df['보통주배당금(천원)']+df['우선주배당금(천원)']
```

```python
df['총자산자연로그']=''
```

```python
# 배당금 지급하면 1을 가지는 더미변수 -> WW지수를 위해

def func_배당금여부(x):
  if x==0:
    return 0
  else:
    return 1
    
```

```python
df['배당금']=df['배당금합산'].map(func_배당금여부)
```

```python
# 총자산에 자연로그를 취한값 -> WW지수를 위해
for i in range(len(df)):
  df['총자산자연로그'][i] = ln(df['자산(*)(요약)'][i])
```

```python
# LEV지수 = 장기차입금 / (총자산 - 자본금 + 시가총액)
df['LEV지수'] = df['장기차입금(요약)'] / (df['자산(*)(요약)'] - df['자본금(요약)'] + df['시가총액'])
```

```python
# 단위 통일 시켜줘야할듯 어떤건 백만원 어떤건 천원이럼 (단위통일 필수 !!!! )
# 자산(*)(요약) -> (*) 총이라는 뜻인듯!
# KZ지수 = = -1.002 * (당기순이익 + 감가상각비) + 0.28 (시가총액 + 순자산가치)/(장기차입금 + 자본)+3.139 * (장기차입금/장기차입금+자본금) - 39.368(보통주배당금+우선주배당금/총자산)-1.315(현금과 현금성자산/총자산)
# 순자산가치 = 자기자본
df['KZ지수'] = -1.002 * (df['당기순이익(손실)(요약)'] + df['감가상각비(백만원)']*1000000) + 0.28* (df['시가총액'] + 순자산가치) / (df['장기차입금(요약)']+df['자본금(요약)']) + 3.139 *(df['장기차입금(요약)']/(df['장기차입금(요약)']+df['자본금(요약)'])) - 39.368 * ( (df['보통주배당금(천원)']*1000 +df['우선주배당금(천원)'*1000])/df['자산(*)(요약)']) -1.315(df['현금및현금성자산(요약)']/df['자산(*)(요약)'])
```

```python
# WW지수 = -0.091(현금흐름 / 총자산) - 0.062*(배당금을 지급하면 1을 가지는 더미변수) +0.021 * (장기차입금/총자산) - 0.044 * (총자산에 자연로그를 취한값) + 0.102 * (기업이 속한 산업의 매출성장률)+0.035 * (기업의 매출성장률)
# 현금흐름은 영업에서 창출된 현금흐름
df['WW지수'] = -0.091 * (df['*영업에서창출된현금흐름(요약)']/df['자산(*)(요약)']) - 0.062 * df['배당금'] + 0.021 (df['장기차입금(요약)']/df['자산(*)(요약)']) - 0.044 * df['총자산자연로그'] + 0.102 * (기업이 속한 산업의 매출성장률) + 0.035 * (기업의 매출성장률)
```

```python
# CL지수(FCCL) = = 0.0432 * (유동자산 / 유동부채) - 0.0011 * [(법인세비용차감전순이익-이자비용)/(이자비용+우선주배당금)] + 0.0039 * [(현금+0.5재고자산 +0.7매출채권-단기차입금)/유형자산]-3.3525*(당기순이익 / 총매출) -0.5723 * (매출성장률=매출액증가율) +0.6067 *(장기차입금/총자산)
df['CL지수'] = 0.0432 * (df['유동자산(*)(요약)']/df['유동부채(*)(요약)']) - 0.0011 * [(df['법인세비용차감전순이익(손실)(요약)']-df['*이자비용(요약)'])/(df['*이자비용(요약)']+df['우선주배당금(천원)'])] + 0.0039 *[(df['현금및현금성자산(요약)']+0.5*df['재고자산(요약)']+0.7*df['매출채권및기타유동채권(요약)']-df['단기차입금(요약)'])/df['유형자산(요약)'])-3.3525*(df['당기순이익(손실)(요약)']/df['매출액(수익)(요약)']) -0.5723 * df['매출액증가율'] +0.6067*(df['장기차입금(요약)']/df['자산(*)(요약)'])
```
